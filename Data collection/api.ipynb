```python
# pip install requests
import requests
```

```python
url="https://stephen-king-api.onrender.com/api/books"

# get type specialy use for getting the data from the api
res=requests.get(url)
```

```python
res
# <Response [200]> is a status code
```

```python
# store the json data into json_Data
json_data=res.json()
```

```python
import pandas as pd

# convert the json data to panda dataframe
df=pd.json_normalize(json_data["data"])
df=df[['id','Title','Year','ISBN']]
```
