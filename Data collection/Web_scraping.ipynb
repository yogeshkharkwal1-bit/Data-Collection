```python
# single page scraping
with open("scraped_data/data2.html","w",encoding="utf8") as f:
    f.write(res.text)
```

```python
# Beautifulsoop
# pip install beautifulsoup
from bs4 import BeautifulSoup
# with open("scraped_data/data2.html","r") as f:
#     html_content=f.read()

# soup=BeautifulSoup(html_content,"lxml")    
with open("scraped_data/data2.html","r") as f:
    html_content=f.read()
soup=BeautifulSoup(html_content,"lxml")    
```

```python
# soup
```

```python
# how to get useful data from the web
# soup.find("h3") # find krta he h3 ko
# soup.h3 #dono same kaam krte he

all_h3=soup.find_all("h3")

all_data=[]

type(all_h3)
for h3 in all_h3:
    country_name=h3.get_text(strip=True)
    # print(h3.find_parent("div")["class"]) that find the parent 
    # print(h3.find_next("div").select("span.country-population"))
    # print(h3.find_next("div").select_one("span.country-population").get_text(strip=True))
    
    population=h3.find_next("div").select("span.country-population")[0].get_text(strip=True)

    all_data.append([country_name,population])
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(all_data,columns=["Name","Population"])
df
```
