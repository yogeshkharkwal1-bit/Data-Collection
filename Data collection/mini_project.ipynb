```python
import requests
from bs4 import BeautifulSoup
```

```python
page_count=1
while True:  
    url=f"https://quotes.toscrape.com/page/{page_count}/"
    res=requests.get(url)
    
    soup=BeautifulSoup(res.text,"lxml")
    quotes=soup.select("div.quote")

    if not quotes:
        print("no valid pages anymore")
        break                               

    with open(f"assignment/quotes{page_count}.html","w",encoding="utf8") as f:    
        f.write(res.text)
        print(f"dowloded data from page{page_count}")

    page_count+=1
 
```

```python
life_quotes=[]
for i in range(1,11):
    with open(f"assignment/quotes{i}.html","r",encoding="utf8") as f:
        html_content=f.read()
          
    # soup=BeautifulSoup(html_content,"lxml")
    soup=BeautifulSoup(html_content,"lxml")
    
    all_quotes=soup.select("div.quote")

    for q in all_quotes:
        all_tags=[]
        
        for tag in q.select(".tag"):
            all_tags.append(tag.get_text().lower())

        if "life" in all_tags:
            text=q.select_one("span.text").get_text(strip=True)
            auther=q.select_one("small.author").get_text(strip=True)

            life_quotes.append([text,auther])

        
print(life_quotes)  

# soup
# all_quotes
# all_tags
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(life_quotes,columns=["Thoughts","Authers"])
```
